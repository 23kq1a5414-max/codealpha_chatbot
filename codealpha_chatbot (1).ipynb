```python
import random

# Global variable to store user's name
user_name = None

def chatbot_response(user_input):
    global user_name
    user_input = user_input.lower()

    if user_input == "hello":
        if user_name:
            return f"Hi {user_name}!"
        else:
            return "Hi there! What's your name?"
    elif user_input.startswith("my name is"):
        # Extract the name after "my name is"
        user_name = user_input.replace("my name is", "").strip().capitalize()
        return f"Nice to meet you, {user_name}!"
    elif user_input == "how are you":
        return "I'm fine, thanks! How about you?"
    elif user_input == "bye":
        if user_name:
            return f"Goodbye {user_name}, have a great day!"
        else:
            return "Goodbye! Have a great day!"
    elif user_input == "what's your name":
        return "I'm a simple chatbot created to chat with you."
    elif user_input == "tell me a joke":
        jokes = [
            "Why don’t scientists trust atoms? Because they make up everything!",
            "Why did the computer go to the doctor? Because it caught a virus!",
            "Why was the math book sad? Because it had too many problems."
        ]
        return random.choice(jokes)
    elif user_input == "help":
        return "You can say things like 'hello', 'how are you', 'bye', 'my name is ...', 'what's your name', or 'tell me a joke'."
    else:
        return "Hmm, I don't understand that. Try 'help' to see what I can do."

def run_chatbot():
    print("Chatbot is running... Type 'bye' to exit.")
    while True:
        user_input = input("You: ")
        response = chatbot_response(user_input)
        print("Bot:", response)
        if user_input.lower() == "bye":
            break

# Start the chatbot
run_chatbot()
```
